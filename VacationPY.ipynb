```python
temp = []
humid = []
cloud = []
wind = []
lat = []
dtime = []
country = []
long=[]
city=[]

for i in df3['City']:
    a = requests.get(url+i).json()
    try:
        print(a['name'])
        temp.append(a['main']['temp'])
        humid.append(a['main']['humidity'])
        wind.append(a['wind']['speed'])
        cloud.append(a['clouds']['all'])
        lat.append(a['coord']['lat'])
        dtime.append(a['dt'])
        country.append(a['sys']['country'])
        long.append(a['coord']['lon'])
        city.append(a['name'])
    except:
        print("n/a")
        pass
```
